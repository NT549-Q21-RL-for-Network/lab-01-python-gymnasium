````python
# Summary of Python Basics and some syntax examples

def section(title):
  print("\n" + "="*len(title))
  print(title)
  print("="*len(title))

# 1. Lists
section("Lists")
fruits = ["apple", "banana", "cherry"]
print("Initial list:", fruits)

# common operations
fruits.append("date")
print("After append:", fruits)
fruits.insert(1, "blueberry")
print("After insert at index 1:", fruits)
print("Slice [1:3]:", fruits[1:3])
print("List comprehension (uppercased):", [f.upper() for f in fruits])

# 2. Tuples
section("Tuples")
point = (10, 20)
print("Tuple (immutable):", point)
# unpacking
x, y = point
print("Unpacked x, y:", x, y)
# single-element tuple
single = (42,)
print("Single-element tuple:", single)

# 3. Dictionaries
section("Dictionaries")
student = {"name": "Alice", "age": 23, "major": "CS"}
print("Dictionary:", student)
print("Access value by key:", student["name"])
student["gpa"] = 3.8
print("After adding key:", student)
print("Iterate keys and values:")
for k, v in student.items():
  print(" ", k, "->", v)

# 4. Sets
section("Sets")
a = {1, 2, 3}
b = {2, 3, 4}
print("Set A:", a)
print("Set B:", b)
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference A - B:", a - b)

# 5. Control flow: if / for / while
section("Control Flow")

n = 7
if n % 2 == 0:
  print(n, "is even")
else:
  print(n, "is odd")

print("For loop over list:")
for idx, fruit in enumerate(fruits):
  print(" ", idx, fruit)

print("While loop (count down):")
count = 3
while count > 0:
  print(" ", count)
  count -= 1
print(" 0 (end)")

# 6. Functions
section("Functions")

def greet(name="Guest"):
  """Return a greeting string."""
  return f"Hello, {name}!"

print(greet("Phat"))
print(greet())

def sum_and_product(a, b):
  """Return multiple values: sum and product."""
  return a + b, a * b

s, p = sum_and_product(3, 5)
print("sum:", s, "product:", p)

# 7. Putting it together: a small example function
section("Mini Project Example: summarize a list of numbers")

def summarize_numbers(nums):
  """
  Given an iterable of numbers, return a dictionary summary:
  - count, sum, min, max, average
  - unique values (as a set)
  """
  nums = list(nums)  # allow any iterable
  if not nums:
    return {"count": 0}
  summary = {
    "count": len(nums),
    "sum": sum(nums),
    "min": min(nums),
    "max": max(nums),
    "average": sum(nums) / len(nums),
    "unique": set(nums)
  }
  return summary

numbers = [1, 2, 2, 3, 4, 4, 4]
print("Numbers:", numbers)
print("Summary:", summarize_numbers(numbers))
````

## Part 2. Object-Oriented Programming (OOP) with Python

````python
# Example of a simple class to represent a network router
# This is a basic implementation to demonstrate object-oriented programming (OOP) concepts in Python.

class Router:
    def __init__(self, name, ip_address):
        self.name = name
        self.ip_address = ip_address
        self.connected_devices = []

    def connect_device(self, device_name):
        self.connected_devices.append(device_name)
        print(f"{device_name} has been connected to {self.name}.")

    def disconnect_device(self, device_name):
        if device_name in self.connected_devices:
            self.connected_devices.remove(device_name)
            print(f"{device_name} has been disconnected from {self.name}.")
        else:
            print(f"{device_name} is not connected to {self.name}.")

    def show_connected_devices(self):
        print(f"Devices connected to {self.name}:")
        for device in self.connected_devices:
            print(f"- {device}")
# Example usage
router1 = Router("Home Router", "192.168.1.1")
router1.connect_device("Laptop")
router1.connect_device("Smartphone")
router1.show_connected_devices()
router1.disconnect_device("Laptop")
router1.show_connected_devices()
````

#### **Exercise 7**: Write a Function (Leap Year) (10) - Phat NT

**Background**

An extra day is added to the calendar almost every four years as February 29, and the day is called a leap day. It corrects the calendar for the fact that our planet takes approximately 365.25 days to orbit the sun. A leap year contains a leap day.

In the Gregorian calendar, three conditions are used to identify leap years:

- The year can be evenly divided by 4, is a leap year, unless:
- The year can be evenly divided by 100, it is NOT a leap year, unless:
- The year is also evenly divisible by 400. Then it is a leap year.

This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years.

**Task**

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean `True`, otherwise return `False`.

Note that the code stub provided reads from STDIN and passes arguments to the `is_leap` function. It is only necessary to complete the `is_leap` function.

**Input Format**

Read `year`, the year to test.

**Constraints**

`1900 <= year <= 10^5`

**Output Format**

The function must return a Boolean value (True/False). Output is handled by the provided code stub.

**Sample Input 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
1990
</pre>
</div>

**Sample Output 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
False
</pre>
</div>

**Explanation 0**

1990 is not a multiple of 4 hence it's not a leap year.

Idea: a year is leap if it is divisible by 4 but not by 100, or if it is divisible by 400; otherwise it is not a leap year.

````python
def is_leap(year):
    leap = False
    if ( year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
````

#### **Exercise 10**: Arrays (NumPy) (20) - Phat NT

The NumPy (Numeric Python) package helps us manipulate large arrays and matrices of numeric data.

To use the NumPy module, we need to import it using:

```text
import numpy
```

**Arrays**

A NumPy array is a grid of values. They are similar to lists, except that every element of an array must be the same type.

```text
import numpy

a = numpy.array([1,2,3,4,5])
print a[1]          #2

b = numpy.array([1,2,3,4,5],float)
print b[1]          #2.0
```

In the above example, `numpy.array()` is used to convert a list into a NumPy array. The second argument (`float`) can be used to set the type of array elements.

**Task**

You are given a space separated list of numbers.

Your task is to print a reversed NumPy array with the element type `float`.

**Input Format**

A single line of input containing space separated numbers.

**Output Format**

Print the reverse NumPy array with type `float`.

**Sample Input**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
1 2 3 4 -8 -10
</pre>
</div>

**Sample Output**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
[-10.  -8.   4.   3.   2.   1.]
</pre>
</div>

Idea: split input into a list, convert to `numpy.array(..., float)`, reverse with `[::-1]`, then print.

````python
import numpy

def arrays(arr):
    return numpy.array(arr, float)[::-1]

arr = input().strip().split()
result = arrays(arr)
print(result)
````
