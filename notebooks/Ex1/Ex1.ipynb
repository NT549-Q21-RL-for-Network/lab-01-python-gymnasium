## Part 2. Object-Oriented Programming (OOP) with Python

````python
# Example of a simple class to represent a network router
# This is a basic implementation to demonstrate object-oriented programming (OOP) concepts in Python.

class Router:
    def __init__(self, name, ip_address):
        self.name = name
        self.ip_address = ip_address
        self.connected_devices = []

    def connect_device(self, device_name):
        self.connected_devices.append(device_name)
        print(f"{device_name} has been connected to {self.name}.")

    def disconnect_device(self, device_name):
        if device_name in self.connected_devices:
            self.connected_devices.remove(device_name)
            print(f"{device_name} has been disconnected from {self.name}.")
        else:
            print(f"{device_name} is not connected to {self.name}.")

    def show_connected_devices(self):
        print(f"Devices connected to {self.name}:")
        for device in self.connected_devices:
            print(f"- {device}")
# Example usage
router1 = Router("Home Router", "192.168.1.1")
router1.connect_device("Laptop")
router1.connect_device("Smartphone")
router1.show_connected_devices()
router1.disconnect_device("Laptop")
router1.show_connected_devices()
````

#### **Exercise 7**: Write a Function (Leap Year) (10) - Phat NT

**Background**

An extra day is added to the calendar almost every four years as February 29, and the day is called a leap day. It corrects the calendar for the fact that our planet takes approximately 365.25 days to orbit the sun. A leap year contains a leap day.

In the Gregorian calendar, three conditions are used to identify leap years:

- The year can be evenly divided by 4, is a leap year, unless:
- The year can be evenly divided by 100, it is NOT a leap year, unless:
- The year is also evenly divisible by 400. Then it is a leap year.

This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years.

**Task**

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean `True`, otherwise return `False`.

Note that the code stub provided reads from STDIN and passes arguments to the `is_leap` function. It is only necessary to complete the `is_leap` function.

**Input Format**

Read `year`, the year to test.

**Constraints**

`1900 <= year <= 10^5`

**Output Format**

The function must return a Boolean value (True/False). Output is handled by the provided code stub.

**Sample Input 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
1990
</pre>
</div>

**Sample Output 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
False
</pre>
</div>

**Explanation 0**

1990 is not a multiple of 4 hence it's not a leap year.

Idea: a year is leap if it is divisible by 4 but not by 100, or if it is divisible by 400; otherwise it is not a leap year.

````python
def is_leap(year):
    leap = False
    if ( year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
````

#### **Exercise 15**: Nested Lists (10) - KienLT

Description: Given the names and grades for each student in a class of `N` students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.

✅ Example

5
Harry
37.21
Berry
37.21
Tina
37.2
Akriti
41
Harsh
39

The answer is: Berry
Harry

Method of implementation, algorithm and approach to solve the problem.

````python
# Ý tưởng: Ta sort theo thứ tự tăng dần của value và lưu vào 1 list, dùng set để lưu các phần tử riêng biệt. Sau đó gán giá trị của phần tử thứ 2 cho 1 biến. Chạy for các items của dict ban đầu, so sánh value vs biến vừa lưu, nếu thỏa thì thêm vào 1 mảng "names". Ta sort mảng "names" theo thứ tự tăng dần và xuất ra từng phần tử trong mảng đó.
````

````python
if __name__ == '__main__':
    students = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        students.append((name, score))

    scores = sorted(set([s[1] for s in students]))
    second_lowest = scores[1]
    
    names = [s[0] for s in students if s[1] == second_lowest]
    names.sort()
    for x in names:
        print(x)
            
````

    - `beta`: [30, 50, 70]
    - query_name = `beta`
- The `query_name` = `beta`. beta's average score is (30 + 50 + 70)/3 = 50.0 

Method of implementation, algorithm and approach to solve the problem.

````python
# Ý tưởng: Tạo 1 biến tổng điểm, và chạy for cộng điểm vào biến đó. Sau đó chia tổng điểm cho số lượng điểm của người đó. Dùng f string để định dạng kết quả đầu ra với 2 chữ số thập phân.
````

````python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
    sumMarks = 0
    for x in student_marks[query_name]:
        sumMarks += x
    cntMarks = len(student_marks[query_name])
    avgMarks = sumMarks/cntMarks
    print(f"{avgMarks:.2f}")
````

#### **Exercise 19**: sWAP cASE (10) - KienLT

Description: You are given a string and your task is to swap cases. In other words, convert all lowercase letters to uppercase letters and vice versa.

**For Example**:
```
Www.HackerRank.com → wWW.hACKERrANK.COM
Pythonist 2 → pYTHONIST 2  
```

**Function Description**
Complete the swap_case function in the editor below.
swap_case has the following parameters:
- string s: the string to modify

Returns:
- string: the modified string

Method of implementation, algorithm and approach to solve the problem.

````python
# Ý tưởng: Tạo 1 chuỗi rỗng là kq trả về. So sánh từng kí tự trong chuỗi truyền vào, nếu trong khoảng "a" đến "z" thì cộng kí tự viết hoa của chữ đó. Nếu trong khoảng "A" đến "Z" thì cộng kí tự viết thường của chữ đó. Các trường hợp khác đều cộng vào chuỗi kq. 
````

````python
def swap_case(s):
    ans = ""
    for i in s:
        if 'a' <= i and i <= 'z':
            ans += i.upper()
        elif 'A' <= i and i <= 'Z':
            ans += i.lower()
        else:
            ans += i
    return ans

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
````

#### **Exercise 22**: Find a String (10) - KienLT

Description: In this challenge, the user enters a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.

**NOTE**: String letters are case-sensitive.

Method of implementation, algorithm and approach to solve the problem.

````python
# Ý tưởng: Duyệt 2 vòng for lồng nhau để kiểm tra. Vòng for ngoài duyệt từng phần tử của chuỗi string, đến khi có kí tự giống với kí tự đầu tiên của chuỗi sub_string thì vòng for trong sẽ kiểm tra xem chuỗi sub_string có khớp với phần tiếp theo của chuỗi string hay không. Nếu khớp thì biến count sẽ tăng lên 1. 
````

````python
def count_substring(string, sub_string):
    cnt = 0
    for i in range(len(string)):
        if string[i] == sub_string[0]:
            tmp = 0
            for j in range(len(sub_string)):
                if i + j < len(string):
                    if string[i+j] == sub_string[j]:
                        tmp += 1
            if tmp == len(sub_string):
                cnt += 1
    return cnt

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
````
