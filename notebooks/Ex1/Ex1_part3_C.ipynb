#### **Exercise 3**: Write a Function (Leap Year) (10)

**Background**

An extra day is added to the calendar almost every four years as February 29, and the day is called a leap day. It corrects the calendar for the fact that our planet takes approximately 365.25 days to orbit the sun. A leap year contains a leap day.

In the Gregorian calendar, three conditions are used to identify leap years:

- The year can be evenly divided by 4, is a leap year, unless:
- The year can be evenly divided by 100, it is NOT a leap year, unless:
- The year is also evenly divisible by 400. Then it is a leap year.

This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years.

**Task**

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean `True`, otherwise return `False`.

Note that the code stub provided reads from STDIN and passes arguments to the `is_leap` function. It is only necessary to complete the `is_leap` function.

**Input Format**

Read `year`, the year to test.

**Constraints**

`1900 <= year <= 10^5`

**Output Format**

The function must return a Boolean value (True/False). Output is handled by the provided code stub.

**Sample Input 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
1990
</pre>
</div>

**Sample Output 0**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
False
</pre>
</div>

**Explanation 0**

1990 is not a multiple of 4 hence it's not a leap year.

Idea: a year is leap if it is divisible by 4 but not by 100, or if it is divisible by 400; otherwise it is not a leap year.

````python
def is_leap(year):
    leap = False
    if ( year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
````

#### **Exercise 4**: Print Function (20)

**Task**

**Example**


#### **Exercise 6**: Arrays (NumPy) (20)

The NumPy (Numeric Python) package helps us manipulate large arrays and matrices of numeric data.

To use the NumPy module, we need to import it using:

```text
import numpy
```

**Arrays**

A NumPy array is a grid of values. They are similar to lists, except that every element of an array must be the same type.

```text
import numpy

a = numpy.array([1,2,3,4,5])
print a[1]          #2

b = numpy.array([1,2,3,4,5],float)
print b[1]          #2.0
```

In the above example, `numpy.array()` is used to convert a list into a NumPy array. The second argument (`float`) can be used to set the type of array elements.

**Task**

You are given a space separated list of numbers.

Your task is to print a reversed NumPy array with the element type `float`.

**Input Format**

A single line of input containing space separated numbers.

**Output Format**

Print the reverse NumPy array with type `float`.

**Sample Input**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
1 2 3 4 -8 -10
</pre>
</div>

**Sample Output**

<div style="border:1px solid #d0d7de; padding:10px 12px; border-radius:6px;">
<pre>
[-10.  -8.   4.   3.   2.   1.]
</pre>
</div>

Idea: split input into a list, convert to `numpy.array(..., float)`, reverse with `[::-1]`, then print.

````python
import numpy

def arrays(arr):
    return numpy.array(arr, float)[::-1]

arr = input().strip().split()
result = arrays(arr)
print(result)
````
