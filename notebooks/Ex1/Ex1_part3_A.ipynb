#### **Exercise 4**: Word Order (50pts)

Description: You are given  words. Some words may repeat. For each word, output its number of occurrences. The output order should correspond with the input order of appearance of the word. See the sample input/output for clarification.

Note: Each input line ends with a "\n" character.

Idea: Dùng Dictionary có thứ tự theo kiểu (từ, value). Từ chưa có sẽ được thêm vào dictionary và cộng 1 vào value của từ đó.

```python
from collections import OrderedDict

n = int(input())
d = OrderedDict()

for _ in range(n):
    word = input()
    d[word] = d.get(word, 0) + 1

print(len(d))
print(*d.values())
```
